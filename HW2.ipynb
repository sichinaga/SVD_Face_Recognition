2. What is the interpretation of the U, Σ and V matrices? (Plot the first few reshaped columns of U)

# U, Σ, V matrix analysis

The columns of the U matrix, where X = UΣV*, respresent the dominant spatial features, or dominant eigenfaces, among all of the images contained within X. The first column of U is the most dominant eigenface, the second column is the second most dominant, and so on. This is due to the fact that the entries along the diagonal of Σ hierarchically decrease in magnitude going down the diagonal. The first six columns of U (and thus the six most dominant eigenfaces) are plotted below.

```python
# Number of columns of U to plot, starting from column 1
num_cols = 6 

# Plot the first num_cols columns of U 
plt.figure(figsize=(8, 6), dpi=100)
for i in range(num_cols):
    plt.subplot(2, 3, i + 1)
    plt.title('u_{}'.format(i + 1))
    plt.imshow(U[:, i].reshape(rows, cols), cmap='gray')
plt.tight_layout()
plt.show()
```

The entries of the Σ matrix, where X = UΣV*, respresent the importance or dominance of their corresponding columns of U and V in representing the data contained within X. For instance, the kth entry of Σ represents the importance of the kth column of U and the kth column of V in representing the images contained within X. This is because the kth entry of Σ is the amount in which values being projected into U space will be scaled in the direction of u_k. The first twenty singular values are plotted below. 

```python
# Number of singular values to plot 
num_plot = 20

# Plot the first num_plot values of Σ
plt.figure(figsize=(5, 3), dpi=100)
plt.title('Singular Values σ_k (first {} values of Σ)'.format(num_plot))
plt.plot(np.arange(1, num_plot + 1), Σ[:num_plot], 'o', c='tab:orange')
plt.xlabel('k')
plt.ylabel('σ_k')
plt.show()
```

The columns of the V matrix, where X = UΣV*, respresent how the images in X project onto the eigenfaces of U scaled by their resepctive value of Σ. For instance, the kth column of V represents how each image in X projects onto the kth column of U, or the kth eigenface, scaled by a factor of σ_k. The first six columns of V (and thus how each image projects onto the six most dominant eigenfaces) are plotted below.

```python
# Number of columns of V to plot, starting from column 1
num_cols = 6 

# Plot the first num_cols columns of V
plt.figure(figsize=(10, 6), dpi=100)
for i in range(num_cols):
    plt.subplot(2, 3, i + 1)
    plt.title('v_{}'.format(i + 1))
    plt.bar(np.arange(num_pics), Vh[i, :])
    plt.xlabel('Image Index')
    plt.ylabel('Projection onto u_{}'.format(i + 1))
plt.tight_layout()
plt.show()
```

3. What does the singular value spectrum look like and how many modes are necessary for good image reconstructions using the PCA basis? (i.e. what is the rank r of the face space?)

# Rank analysis

Since the entries of the Σ matrix, where X = UΣV*, respresent the importance or dominance of their corresponding columns of U and V in representing the data contained within X, we can use the entries of the Σ to determine how much information is retained after a rank r truncation of the matrices of the SVD is used to approximate the images in X. Here, I will utilize the following measure of information retained after a rank r truncation. 
$$
\text{Percent of information retained} = \bigg( \frac{\sum_{i=1}^r \sigma_i^2}{\sum_{i=1}^n \sigma_i^2} \bigg) * 100\%, \quad 1 \leq r \leq n
$$
Under this metric, we can determine that a rank of r = 650 is appropriate for preserving 99.9% of the image information from training and can be used as the rank of the free space, given that 99.9% of the information captured by the SVD during training is desired. 

```python
# Given the Σ matrix from the SVD and the desired rank truncation r, computes 
# and returns the percent of information retained after a rank r approximation. 
def percent_retained(Σ, r): 
    return (np.sum(Σ[:r] ** 2) / np.sum(Σ ** 2)) * 100

# Given an image as a vector, the U matrix from the SVD, and the desired 
# rank r truncation, computes and returns a rank r approximation of the 
# given image 
def reconstruct(img, U, r):
    Ur = U[:, :r]
    β = Ur.T @ img
    recon = Ur @ β
    return recon
```

```python
# Choose the number of images to use for model training 
num_train = int(0.7 * num_pics)
num_valid = num_pics - num_train
X_valid = X[:, :num_valid] # Validation set 
X_train = X[:, num_valid:] # Training set 

# Perform the SVD again, but only on the training data 
Ut, Σt, Vht = np.linalg.svd(X_train, full_matrices=False)

# Plot all of the singular values
plt.figure(figsize=(8, 3), dpi=100)
plt.title('Singular values from training data')
plt.plot(np.arange(1, len(Σt) + 1), Σt, '.', c='tab:orange')
plt.xlabel('k')
plt.ylabel('σ_k')
plt.show()
```

```python
# Total number of singular values 
n = len(Σt)

# Compute the percent of information in X retained by each rank r truncation
ranks = np.arange(1, n + 1)
percents = np.empty(n)
for i, r in enumerate(ranks):
    percents[i] = percent_retained(Σt, r)

# Find the minimum rank truncation needed to retain the following percentages of information
percent = [99.9, 99, 98, 95]
rank = []
for percentage in percent:
    r = 1
    info_retained = percent_retained(Σt, r)
    while (info_retained < percentage):
        r += 1
        info_retained = percent_retained(Σt, r)
    rank.append(r)
    
# Image for testing image reconstruction 
test_img = X_valid[:, 0]

# Plot results 
plt.figure(figsize=(15, 5), dpi=100)
plt.suptitle('Rank reconstructions', fontsize=20)

plt.subplot(2, 5, 1)
plt.title('Percent of information retained')
plt.plot(ranks, percents, '.', c='tab:red')
plt.xlabel('Rank truncation r')
plt.ylabel('Percent')

plt.subplot(2, 5, len(percent) + 2)
plt.title('Original image (Rank {})'.format(len(Σt)))
plt.imshow(test_img.reshape(rows, cols), cmap='gray')
plt.xticks([])
plt.yticks([])

for i in range(len(percent)):
    p = percent[i]
    r = rank[i]
    
    plt.subplot(2, 5, i + 2)
    plt.title('{}% information retained'.format(p))
    plt.plot(ranks, percents, '.', c='tab:orange')
    plt.plot(ranks[:r], percents[:r], 'o', c='tab:red')
    plt.xlabel('Rank truncation r')
    plt.ylabel('Percent')

    plt.subplot(2, 5, i + len(percent) + 3)
    plt.title('Rank {} reconstruction'.format(r))
    plt.imshow(reconstruct(test_img, Ut, r).reshape(rows, cols), cmap='gray')
    plt.xticks([])
    plt.yticks([])
    
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
